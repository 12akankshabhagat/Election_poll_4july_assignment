## EDA

```python
# Plot a histogram for the age distribution of voters

plt.figure(figsize=(8, 5))
sns.histplot(df['age'],  color='blue')
plt.title('Age Distribution of Voters')
plt.xlabel('Age')
plt.ylabel("count")
plt.show()
```

### Insight:
#### From 35 to 43 age groups people is most engaged in election

```python
df["sex"].value_counts()
```

```python
#pie chart
plt.pie(df["sex"].value_counts(),labels=["Male","Female"],autopct="%0.2f%%")
plt.show()
     
```

#### male are more active to donate vote as campare to female.

```python
# this number of people knows about election date

plt.figure(figsize=(10,5))
df.knows_election_date.value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Percentage of people who knows election date")
plt.legend()
plt.show()
```

#### 85.1% people knows election date

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['will_vote'],  color='blue')
plt.title('this are no. of people will vote')
plt.xlabel('vote')
plt.ylabel("count")
plt.show()
```

#### More no. of people will vote

```python
#### Count of Sex with Respect to Federal Districts

print("federal district count is :",len(df["federal_district"].unique()))
```

```python
custom_palette = [ "blue" , "orange"]  
plt.figure(figsize=(10, 7))
ax = sns.countplot(data=df, x="federal_district", palette=custom_palette, hue="sex")  
plt.xlabel("Federal District")
plt.ylabel("Count")
plt.title("Count of Sex with Respect to Federal Districts")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

#### total 8 federal district in election
#### the highest count of male and female in center federal district.
#### the lowest count of male and female in far eastern

```python
## which types of city has more votes?
 
plt.figure(figsize=(10,5))
sns.countplot(data=df,x="type_of_city",palette="magma",hue="sex")
plt.xlabel("Type of city ,population")
plt.ylabel("count")
plt.title("count of type of city with population")
plt.xticks(rotation=90)
plt.show()
```

#### There are 7 city is going do election.

#### higher voter in village.

#### less no. of voter in settlement

#### female vote are more as campare to male vote in gives city column


```python
# this are candidate who stand for election

df.candidate.unique()
```

```python
df['candidate'].value_counts().plot.bar(color='Silver')
plt.xlabel('Candidate')
plt.ylabel('Count')
plt.title('Count of Votes for Each Candidate')
plt.xticks(rotation=45)
plt.grid(axis='y')  # Adding horizontal grid lines
plt.show()
```

#### candidate name putin has recive more vote and chance to win.

```python
# what are the educational background of voter

df_educated_voters =df.education.value_counts()
```

```python
plt.figure(figsize=(9,7))
sns.barplot(x=df_educated_voters.index,y=df_educated_voters.values,palette="magma")
plt.xticks(rotation =45)
plt.title("Education of voters")
plt.show()
```

#### from above barplot is observed that college students are more active for voting.
